### Matplotlib로 기본 챠트 그리기

```python
# numpy를 이용한 sine 그래프
import numpy as np

t = np.arange(0,12,0.01) #0부터 12까지 0.01 간격으로 데이타 만들기 
print(t)
y = np.sin(t)
print(y)
```

### 캔버스 객체 생성

```python
## 캔버스 객체 생성
fig = plt.figure(figsize=(6,4)) # 캔버스 객체 생성(사이즈를 figsize에 할당)
```

```python
## axes 축에 대한 굵기를 진하게 처리한다
plt.rcParams['axes.linewidth'] = 2.0      
```

```python
fig = plt.figure(figsize=(6,4))
plt.plot(t, y)
plt.show()
```

### 그리드 적용하기

```python
plt.figure(figsize=(6,4))
plt.plot(t, y)
plt.grid() # 그리드 적용하기 
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(t, y)
plt.grid()
plt.xlabel('time')       # x축 라벨 적용하기
plt.ylabel('Amplitude')  # y축 라벨 적용하기
plt.show()
```

### x축 라벨, y축 제목 달기

```python
plt.figure(figsize=(6,4))
plt.plot(t, y)
plt.grid()
plt.xlabel('time')      # x축 라벨
plt.ylabel('Amplitude') # y축 라벨
plt.title('Example of sinewave') # 그래프의 제목 
plt.show()
```

### 2개 그래프 그리기

```python
plt.figure(figsize=(6,4))
plt.plot(t, np.sin(t))
plt.plot(t, np.cos(t)) # 그래프 추가
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

### legend(범례) 표시

```python
plt.figure(figsize=(6,4))
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.grid()
plt.legend() # 범례 표시 
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

### 선의 굵기와 색상 지정

```python
plt.figure(figsize=(6,4))
plt.plot(t, np.sin(t), lw=3, label='sin') # lw : 선의 굵기 조정 
plt.plot(t, np.cos(t), color='r', label='cos')  #  color : 색상 지정
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

### x축과 y축의 범위 지정

```python
plt.figure(figsize=(6,4))
plt.plot(t, np.sin(t), lw=3, label='sin')
plt.plot(t, np.cos(t), 'r', label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.ylim(-1.2, 1.2) # y축의 범위 지정
plt.xlim(0, np.pi)  # x축의 범위 지정
plt.show()
```

### 한글 폰트 지정

```python
import sys                                               ## 파이썬 엔진에 대한 정보를 관리하는 모듈을 사용한다. 
from matplotlib import font_manager, rc                  ## 폰트를 관리하는 함수와 설정 함수를 사용한다. 

if sys.platform  == 'darwin':                             ## MAC OS의 이름을 확인한다.
    path = '/Library/Fonts/AppleGothic.ttf'  
elif sys.platform == 'win32':                             ## Windows 이름을 확인한다.
    path = "c:/Windows/Fonts/malgun.ttf"
else:
    print('Unknown system... sorry~~~~') 
    
font_name = font_manager.FontProperties(fname=path).get_name()        ##  폰트가 있는지를 확인한다. 
rc('font', family=font_name)                                          ## 한글 폰트를 시각화 환경에 세팅한다. 
plt.rcParams['axes.unicode_minus'] = False                            ## 한글 처리시 마이너스 값을 처리할 수 있도록 조정한다.
```

### 그래프 좌표에 대한 레이블 지정

```python
x = np.arange(1,10)                  ## 9 개의 원소를 가진 1차원 배열을 만든다.
y = x **2                            ## 이 배열을 제곱한  1차원 배열을 만든다.

fig = plt.figure(figsize=(8,4))
plt.plot(x,y, c="r", lw=3,ls="--", marker="s", ms=6, mec="b", mew=3, mfc="r")
plt.title("x축과 y축의 tick설정")
plt.xticks(x)                    ## 좌표 내의 숫자의 값을 지정한다. 
plt.yticks(y)
plt.show()
```

### 서브 플롯 추가하기

```python
# data
t = [0, 1, 2, 3, 4, 5, 6] 
y = [1, 4, 5, 8, 9, 5, 3]

# graph 
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_subplot(2, 2, 1) # 2행 2열 그래프, 1번째 그래프
ax1.plot(t, y, color='green') 

ax2 = fig.add_subplot(2, 2, 2) # 2행 2열 그래프, 2번째 그래프
ax2.plot(t, y, color='green', linestyle='dashed')  

ax3 = fig.add_subplot(2, 2, 3) # 2행 2열 그래프, 2번째 그래프
ax3.plot(t, y, color='green', linestyle='dashed', marker='o') 

ax4 = fig.add_subplot(2, 2, 4) # 2행 2열 그래프, 4번째 그래프
ax4.plot(t, y, color='green', linestyle='dashed', marker='o', markerfacecolor = 'blue')

plt.show()
```

### linestyle, marker 지정

```python
plt.figure(figsize=(10,6))
plt.plot(t, y, color='green', linestyle='dashed') # linestyle='dashed' 옵션 - 선 스타일 지정 
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, y, color='green', linestyle='dashed', marker='o') # marker 옵션 - 데이타가 존재하는 곳에 마킹
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, y, color='green', linestyle='dashed', marker='o',
        markerfacecolor = 'blue') # markerfacecolor, markersize 옵션 -> 마커의 크기와 색상 지정
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, y, color='green', linestyle='dashed', marker='o',
        markerfacecolor = 'blue', markersize=12)

plt.xlim([-0.5, 6.5])
plt.ylim([0.5, 9.5])
plt.show()
```

### color map

```python
colormap = t

plt.figure(figsize=(10,6))
plt.scatter(t,y, s = 50, c = colormap, marker='>') 
# x 축 값인 t에 따라 색상을 바꾸는 color map 지정, s 옵션 : 마커의 크기
plt.colorbar()
plt.show()
```

```python
# numpy 랜덤변수 함수를 이용하여 3개의 데이타 생성  
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)
```

```python
plt.figure(figsize=(10,6))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

### boxplot

```python
# boxplot
plt.figure(figsize=(8,4))
plt.boxplot((s1, s2, s3))
plt.grid()
plt.show()
```
